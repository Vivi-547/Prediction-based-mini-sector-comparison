```

import fastf1
import matplotlib.pyplot as plt
import numpy as np

fastf1.Cache.enable_cache('cache')

def predict_performance(lap_data, points):
    factor = 1 - (points / 500)
    return lap_data['Speed'] * factor

def plot_prediction(driver1, driver2, session):
    lap1 = session.laps.pick_driver(driver1).pick_fastest().get_telemetry()
    lap2 = session.laps.pick_driver(driver2).pick_fastest().get_telemetry()

    pred1 = predict_performance(lap1, 400)
    pred2 = predict_performance(lap2, 280)

    fig, ax = plt.subplots(figsize=(15, 8))
    ax.scatter(lap1['X'], lap1['Y'], c=pred1, cmap='Oranges', s=5, label=f'{driver1} Predicted')
    ax.scatter(lap2['X'], lap2['Y'], c=pred2, cmap='Purples', s=5, label=f'{driver2} Predicted')

    ax.legend()
    plt.show()

session = fastf1.get_session(2024, 'Japan', 'R')
session.load()
plot_prediction('VER', 'HAM', session)
```
